```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = ['red', 'green', 'yellow', 'blue']
Rmeas = 100
```

```python
plt.figure()
for color in colors:
    filename = color + '_diode_i_v.csv'
    data = np.loadtxt(filename, skiprows=10, comments=';', delimiter=',')
    # plt.semilogy(data[:,2], data[:,3]/Rmeas*1000, label=color, color=color)
    plt.plot(data[:,2], data[:,3]/Rmeas*1000, label=color, color=color)
plt.xlabel('$V_{Diode}$ [V]')
plt.ylabel('$I_{Diode}$ [mA]')
plt.title('I-V Characteristic LEDs')
plt.legend()
plt.grid(True, which='both')
plt.savefig('I_V_Leds_linear.png')
```

```python
plt.figure()
for color in colors:
    filename = color + '_diode_i_v.csv'
    data = np.loadtxt(filename, skiprows=10, comments=';', delimiter=',')
    plt.semilogy(data[:,2], data[:,3]/Rmeas*1000, label=color, color=color)
plt.xlabel('$V_{Diode}$ [V]')
plt.ylabel('$I_{Diode}$ [mA]')
plt.title('I-V Characteristic LEDs')
plt.legend()
plt.grid(True, which='both')
plt.savefig('I_V_Leds_semilog.png')
```
